```python
import pandas as pd
data1={'Name':['jennie','princi','jimin','jhope'],
        'Age':[27,24,22,32],
        'Address':['mumbai','hyderabad','korea','karnataka'],
        'Qualification':['Msc','MA','MCA','phd'],
        'salary':[1000,2000,3000,4000]}
data2=data1={'Name':['jennie','lisa','suge','felix'],
        'Age':[27,14,12,52],
        'Address':['mumbai','hyderabad','korea','karnataka'],
        'Qualification':['B tech','B com','BSc','B.A'],
        'salary':[1000,2000,3000,4000]}
#to convert the dictionary into dataframe
df=pd.DataFrame(data1,index=[0,1,2,3])
#covert dictionary into dataframe
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
#using a .concat() method
res1=pd.concat([df,df1])
res1
res2=pd.concat([df,df1],axis=1,join='inner')
res2
```

```python
import pandas as pd
data1={'Name':['jennie','princi','jimin','jhope'],
        'Age':[27,24,22,32],
        'Address':['mumbai','hyderabad','korea','karnataka'],
        'Qualification':['Msc','MA','MCA','phd'],
        'salary':[1000,2000,3000,4000]}
data2=data1={'Name':['jennie','lisa','suge','felix'],
        'Age':[27,14,12,52],
        'Address':['mumbai','hyderabad','korea','karnataka'],
        'Qualification':['B tech','B com','BSc','B.A'],
        'salary':[1000,2000,3000,4000]}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
#using ignore index
res=pd.concat([df,df1],ignore_index=True)
res
```
