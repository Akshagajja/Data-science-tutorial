# **# Define a dictionary containing employee data**



```python


import pandas as pd
data1={'Name':['jennie','princi','jimin','jhope'],
        'Age':[27,24,22,32],
        'Address':['mumbai','hyderabad','korea','karnataka'],
        'Qualification':['Msc','MA','MCA','phd'],
        'salary':[1000,2000,3000,4000]}
#convert the dictionary into dataframe
df=pd.DataFrame(data1,index=[0,1,2,3])
#creating a series
s1=pd.Series([1000,2000,3000,4000],name='Salary')
print(df,"/n/n",s1)
#combining series and dataframe
res=pd.concat([df,s1],axis=1)
res
```

```python
data1={'key':['k0','k1','k2','k3'],
       'Name':['jai','princi','jimin','jhope'],
       'Age':[27,24,22,32],}
#define a dictionary containing employee data
data2={'key':['k0','k1','k2','k3'],
       'address':['mumbai','karnataka','jaipur','korea'],
       'Qualification':['btech','B.A','B.C.A','M.S'],}
#convert dictionary into dataframe
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
#using merge()function
res=pd.merge(df,df1,on='key')
res
```
